Searching multiple fields

it mainly gets the relevances score of all the fields and takes the highest, the tie breaker adds the relevance of all the fields 

```python
# Query 1: multi_match for "vegetable" on "name" and "tags" fields
query_multi_match_simple = {
    "query": {"multi_match": {"query": "vegetable", "fields": ["name", "tags"]}}
}

# Query 2: multi_match for "vegetable" with a field boost for "name"
query_multi_match_boosted = {
    "query": {"multi_match": {"query": "vegetable", "fields": ["name^2", "tags"]}}
}

# Query 3: multi_match for "vegetable broth" on "name" and "description" with a tie breaker
query_multi_match_tie_breaker = {
    "query": {
        "multi_match": {
            "query": "vegetable broth",
            "fields": ["name", "description"],
            "tie_breaker": 0.3,
        }
    }
}

# Execute the queries
response_multi_match_simple = es.search(index="products", body=query_multi_match_simple)
response_multi_match_boosted = es.search(
    index="products", body=query_multi_match_boosted
)
response_multi_match_tie_breaker = es.search(
    index="products", body=query_multi_match_tie_breaker
)

# Print the responses
print("Response for multi_match on 'name' and 'tags':", response_multi_match_simple)
print("Response for multi_match with 'name' boosted:", response_multi_match_boosted)
print("Response for multi_match with 'tie_breaker':", response_multi_match_tie_breaker)
```

if a bool query only contains match clauses, at least one of them must match but if a must query exist the should only boost the relevance scores

filter is similar to must but it disregards the relevance score

```python
# Query 1: Must contain "Alcohol" in tags.keyword
query_1 = {"query": {"bool": {"must": [{"term": {"tags.keyword": "Alcohol"}}]}}}

# Query 2: Must contain "Alcohol" but must not contain "Wine" in tags.keyword
query_2 = {
    "query": {
        "bool": {
            "must": [{"term": {"tags.keyword": "Alcohol"}}],
            "must_not": [{"term": {"tags.keyword": "Wine"}}],
        }
    }
}

# Query 3: Must contain "Alcohol", must not contain "Wine", and should contain "Beer"
query_3 = {
    "query": {
        "bool": {
            "must": [{"term": {"tags.keyword": "Alcohol"}}],
            "must_not": [{"term": {"tags.keyword": "Wine"}}],
            "should": [{"term": {"tags.keyword": "Beer"}}],
        }
    }
}

# Query 4: Must contain "Alcohol", must not contain "Wine", should contain "Beer" or have "beer" in name or description
query_4 = {
    "query": {
        "bool": {
            "must": [{"term": {"tags.keyword": "Alcohol"}}],
            "must_not": [{"term": {"tags.keyword": "Wine"}}],
            "should": [
                {"term": {"tags.keyword": "Beer"}},
                {"match": {"name": "beer"}},
                {"match": {"description": "beer"}},
            ],
        }
    }
}

# Query 5: Should contain "Beer" in tags.keyword or "beer" in name
query_5 = {
    "query": {
        "bool": {
            "should": [{"term": {"tags.keyword": "Beer"}}, {"match": {"name": "beer"}}]
        }
    }
}

# Query 6: Must contain "Alcohol" in tags.keyword, should contain "Beer" in tags.keyword or "beer" in name
query_6 = {
    "query": {
        "bool": {
            "must": [{"term": {"tags.keyword": "Alcohol"}}],
            "should": [{"term": {"tags.keyword": "Beer"}}, {"match": {"name": "beer"}}],
        }
    }
}

# Query 7: Must contain "Alcohol" in tags.keyword, should contain "Beer" in tags.keyword or "beer" in name, with minimum_should_match
query_7 = {
    "query": {
        "bool": {
            "must": [{"term": {"tags.keyword": "Alcohol"}}],
            "should": [{"term": {"tags.keyword": "Beer"}}, {"match": {"name": "beer"}}],
            "minimum_should_match": 1,
        }
    }
}

# Execute the queries
response_1 = es.search(index="products", body=query_1)
response_2 = es.search(index="products", body=query_2)
response_3 = es.search(index="products", body=query_3)
response_4 = es.search(index="products", body=query_4)
response_5 = es.search(index="products", body=query_5)
response_6 = es.search(index="products", body=query_6)
response_7 = es.search(index="products", body=query_7)

# Print the responses
print("Response for Query 1:", response_1)
print("Response for Query 2:", response_2)
print("Response for Query 3:", response_3)
print("Response for Query 4:", response_4)
print("Response for Query 5:", response_5)
print("Response for Query 6:", response_6)
print("Response for Query 7:", response_7)
```

dis_max

at least one of the queries must match and it takes the highest relevance scores unless tie breaker is added

```python
# 1. Basic disjunction max (dis_max) usage - Queries for "vegetable" in "name" or "tags"
query_1 = {
    "query": {
        "dis_max": {
            "queries": [
                {"match": {"name": "vegetable"}},
                {"match": {"tags": "vegetable"}},
            ]
        }
    }
}

# 2. Disjunction max with a specified tie breaker (0.3)
query_2 = {
    "query": {
        "dis_max": {
            "queries": [
                {"match": {"name": "vegetable"}},
                {"match": {"tags": "vegetable"}},
            ],
            "tie_breaker": 0.3,
        }
    }
}

# Execute the queries
response_1 = es.search(index="products", body=query_1)
response_2 = es.search(index="products", body=query_2)

# Print the responses
print("Response for Query 1:", response_1)
print("Response for Query 2:", response_2)
```

with a nested objects, if nested type is not specified all the child documents will be indexed as 1 which yield strange results, so we have to created a new mapping as type nested so we can query independently the child objects ie ingredients

```python
# Step 1: Incorrect query without nested data type - This won't work as expected because it won't handle arrays of objects correctly
incorrect_query = {
    "query": {
        "bool": {
            "must": [
                {"match": {"ingredients.name": "parmesan"}},
                {"range": {"ingredients.amount": {"gte": 100}}},
            ]
        }
    }
}

# Execute the incorrect query (for reference only; this would not yield correct results in complex cases)
incorrect_response = es.search(index="recipes", body=incorrect_query)
print("Response for Incorrect Query:", incorrect_response)

# Step 2: Delete and create the correct index with nested data type mapping
# Delete the old index if it exists
es.indices.delete(index="recipes", ignore=[400, 404])

# Create a new index with the correct mapping
mapping = {
    "mappings": {
        "properties": {
            "title": {"type": "text"},
            "description": {"type": "text"},
            "preparation_time_minutes": {"type": "integer"},
            "steps": {"type": "text"},
            "created": {"type": "date"},
            "ratings": {"type": "float"},
            "servings": {
                "properties": {"min": {"type": "integer"}, "max": {"type": "integer"}}
            },
            "ingredients": {
                "type": "nested",  # Define ingredients as a nested type
                "properties": {
                    "name": {
                        "type": "text",
                        "fields": {"keyword": {"type": "keyword"}},
                    },
                    "amount": {"type": "integer"},
                    "unit": {"type": "keyword"},
                },
            },
        }
    }
}

# Put the new mapping
es.indices.create(index="recipes", body=mapping)

# Step 3: Correct query using the nested type to search for ingredients with name "parmesan" and amount >= 100
nested_query = {
    "query": {
        "nested": {
            "path": "ingredients",
            "query": {
                "bool": {
                    "must": [
                        {"match": {"ingredients.name": "parmesan"}},
                        {"range": {"ingredients.amount": {"gte": 100}}},
                    ]
                }
            },
        }
    }
}

# Execute the correct nested query
nested_response = es.search(index="recipes", body=nested_query)
print("Response for Nested Query:", nested_response)
```

to know which nested object was matched specify the inner hits

```python
# Query with basic inner hits enabled - Retrieves matching nested objects under "ingredients"
inner_hits_query = {
    "query": {
        "nested": {
            "path": "ingredients",
            "inner_hits": {},  # Enables inner hits without any customization
            "query": {
                "bool": {
                    "must": [
                        {"match": {"ingredients.name": "parmesan"}},
                        {"range": {"ingredients.amount": {"gte": 100}}},
                    ]
                }
            },
        }
    }
}

# Execute the query with inner hits
response_inner_hits = es.search(index="recipes", body=inner_hits_query)
print("Response with Basic Inner Hits:", response_inner_hits)


# Query with customized inner hits - Specifies a custom key and limits the number of hits returned
custom_inner_hits_query = {
    "query": {
        "nested": {
            "path": "ingredients",
            "inner_hits": {
                "name": "my_hits",  # Custom name for the inner hits
                "size": 10,  # Limits the number of inner hits to 10
            },
            "query": {
                "bool": {
                    "must": [
                        {"match": {"ingredients.name": "parmesan"}},
                        {"range": {"ingredients.amount": {"gte": 100}}},
                    ]
                }
            },
        }
    }
}

# Execute the customized inner hits query
response_custom_inner_hits = es.search(index="recipes", body=custom_inner_hits_query)
print("Response with Customized Inner Hits:", response_custom_inner_hits)
```
